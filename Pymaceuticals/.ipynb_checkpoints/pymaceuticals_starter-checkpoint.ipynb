```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_df = pd.merge(clinical_trial_df,mouse_drug_df, on="Mouse ID")

# Display the data table for preview
merged_df.head()
```

## Tumor Response to Treatment

```python
#Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
mean_tumor_volume=merged_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean()

mean_tumor_volume_df = pd.DataFrame(mean_tumor_volume).reset_index()

mean_tumor_volume_df

```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
se_tumor_volume = merged_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()

se_tumor_volume_df = pd.DataFrame(se_tumor_volume).reset_index()

se_tumor_volume_df

```

```python
# Minor Data Munging to Re-Format the Data Frames

mean_tumor_volume_formatted = mean_tumor_volume_df.pivot(
    index='Timepoint',
    columns='Drug', 
    values='Tumor Volume (mm3)')

se_tumor_volume_formatted = se_tumor_volume_df.pivot(
    index = 'Timepoint', 
    columns = 'Drug', 
    values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
se_tumor_volume_formatted.head(5)
mean_tumor_volume_formatted.head(5)

```

```python
# Generate the Plot (with Error Bars)
x = mean_tumor_volume_formatted.index  
loop = mean_tumor_volume_formatted.columns

for i in loop:
    plt.errorbar(x, mean_tumor_volume_formatted[i], yerr = se_tumor_volume_formatted[i], marker = 'o',
                 )
plt.title("Tumour Response to Treatment")
plt.ylabel("Total Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid()
plt.legend(('Capomulin', 'Ceftamin', 'Infubinol', 'Ketapril', 'Naftisol', 
            'Placebo', 'Propriva', 'Ramicane', 'Stelasyn', 'Zoniferol'), 
           loc='center left', bbox_to_anchor=(1, 0.5))


plt.savefig("TumorResponse_Treatment.png")

```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

metsite = merged_df.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame

metsite_df = pd.DataFrame(metsite).reset_index()
# Preview DataFrame

metsite_df.head(5)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

se_metsite = merged_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem()

# Convert to DataFrame
se_metsite_df = pd.DataFrame(se_metsite).reset_index()
# Preview DataFrame

se_metsite_df.head(5)
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_metsite_fmt = metsite_df.pivot(
    index='Timepoint',
    columns='Drug', 
    values='Metastatic Sites')

se_metsite_fmt = se_metsite_df.pivot(
    index = 'Timepoint', 
    columns = 'Drug', 
    values = 'Metastatic Sites')

# Preview that Reformatting worked



se_metsite_fmt.head(5)
mean_metsite_fmt.head(5)
```

```python
# Generate the Plot (with Error Bars)
x = mean_metsite_fmt.index
loop = mean_metsite_fmt.columns

for i in loop: 
    plt.errorbar(x, mean_metsite_fmt[i], yerr=se_metsite_fmt[i], marker = '^')

    
plt.title("Metastatic Response to Treatment")
plt.ylabel("Total Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid()
plt.legend(('Capomulin', 'Ceftamin', 'Infubinol', 'Ketapril', 'Naftisol', 
            'Placebo', 'Propriva', 'Ramicane', 'Stelasyn', 'Zoniferol'), 
           loc='center left', bbox_to_anchor=(1, 0.5))

# Save the Figure
plt.savefig("MetastaticResponse_Treatment.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = merged_df.groupby(['Drug','Timepoint'])['Mouse ID'].count()
# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count).reset_index()
# Preview DataFrame
mice_count_df.head(5)
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_fmt = mice_count_df.pivot(
    index='Timepoint',
    columns='Drug', 
    values='Mouse ID')
# Preview the Data Frame
mice_count_fmt.head(5)
```

```python
# Generate the Plot (Accounting for percentages)

survival_rate = (mice_count_fmt.iloc[:,:]/mice_count_fmt.iloc[0,:]) * 100

x = mice_count_fmt.index
loop = mice_count_fmt.columns

for i in loop: 
    plt.plot(x, survival_rate[i], marker = '*')
    

# Save the Figure
plt.title("Survival Rate to Treatment")
plt.ylabel("Total Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid()
plt.legend(('Capomulin', 'Ceftamin', 'Infubinol', 'Ketapril', 'Naftisol', 
            'Placebo', 'Propriva', 'Ramicane', 'Stelasyn', 'Zoniferol'), 
           loc='best', bbox_to_anchor=(1, 0.5))

# Save the Figure
plt.savefig("SurivalRate.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

percent_change = (mean_tumor_volume_formatted.iloc[9, :] - mean_tumor_volume_formatted.iloc[0,:])/mean_tumor_volume_formatted.iloc[0,:] * 100


# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple

x_axis = np.arange(0,10,1)

ziplist = zip(percent_change.reset_index()['Drug'], percent_change.values)
percent_change_tuples_list = list(ziplist)
```

```python
# Splice the data between passing and failing drugs
drugs = []
colors = []

for i in percent_change_tuples_list: 
    if i[1]<0:
        drugs.append(i), colors.append('g')
    else: 
        drugs.append(i), colors.append('r') 
        
drug_percent = [drugs[i][1] for i in np.arange(len(drugs))]
drug_name = [drugs[i][0] for i in np.arange(len(drugs))]

bar = plt.bar(np.arange(len(drugs)), drug_percent, color = colors, width = 1)

def percent_change (bar, drug_percent): 
    for bar, drug_percent in zip(bar, drug_percent): 
        plt.text(bar.get_x() + bar.get_width()/2, -5, str(f"{drug_percent:.0f}%")
                 )

percent_change(bar,drug_percent)

plt.grid()
plt.title("Tumour % Change throughout 45 Day Treatment")
plt.ylabel("% Tumour Volume Change")
plt.xticks(np.arange(len(percent_change_tuples_list)), drug_name, rotation='65')


plt.savefig("tumour_vol_change_through_45_day_treatment.png")

# Show the Figure
plt.show()
```
